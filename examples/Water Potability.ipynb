```python
import os

os.chdir("../src")

import pandas as pd
from PCA import PCA

from sklearn.impute import KNNImputer
```

```python
data = pd.read_csv("../data/water_potability.csv")
```

```python
knn_imputer = KNNImputer()

X = knn_imputer.fit_transform(data)
```

```python
pca = PCA(X)

pca.fit(ncomps=5)
```

```python
pca.score_plot(1, 2)
```
